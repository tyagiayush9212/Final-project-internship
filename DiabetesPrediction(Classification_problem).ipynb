**Diabetes Prediction - Final Project**


**OBJECTIVE** - The primary objective of this project is to identify the diabetic patients based on several medical factors like sugar level , Blood pressure, pregnancy , etc. using a sample data Diabetes.csv. The main machine learning algorithm used is Logistic Regression and K-Neighbours Classifiers model.


**Data Source** - Main data source is Github.com/YbiFoundationDatasets repository


```python
# Import required libraries
import pandas as pd
```

```python
# Import dataset:
df = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Diabetes.csv');
df.head(10)
```

```python
#Describe data:
df.info()
df.describe()
```

```python
# Selecting target variable and independant variables:
y = df['diabetes']
X = df.drop(['diabetes'], axis=1)
```

```python
df.columns
```

```python
# Split the data:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2529)
```

```python
# Using logistic regression model as it is a classification problem:
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 500)
```

```python
# Training the model
model.fit(X_train , y_train)
```

```python
# Prediction:
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
# Finding the accuracy score:
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```

**Explaination** - The model is now able to classify whether the patient is diabetic or not with an accuracy of 77.08% . This can be increased if other classifiers are used like K-Neighbours Classifier as mentioned .
